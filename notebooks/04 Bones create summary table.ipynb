```python
import pandas as pd
df_transects = pd.read_pickle("../data/pkl/df_transects.pkl")
df_transects.head()
```

```python
df_occurrences = pd.read_pickle("../data/pkl/df_occurrences.pkl")
df_occurrences.head()
```

```python
df_instances = pd.read_pickle("../data/pkl/df_instances.pkl")
df_instances.head()
```

```python
df_briana_with_responses = pd.read_pickle("../data/pkl/df_briana_with_responses.pkl")
df_briana_with_responses.head()
```

```python
df_fire_with_responses = pd.read_pickle("../data/pkl/df_fire_with_responses.pkl")
df_fire_with_responses.head()
```

```python
# Filter transects to only those marked as "on old reserve"
df_old_reserve_transects = df_transects[df_transects["Pre: On old reserve?"] == "Yes"].copy()
valid_transect_uids = df_old_reserve_transects['UID']

# Count by 'Pre: Transect physical habitat'
counts = df_old_reserve_transects["Pre: Transect physical habitat"].value_counts()

# Convert to a single-row DataFrame with habitat types as columns
summary_table = pd.DataFrame([counts])
summary_table.index = ['# of transects']

# Step 2: Join df_occurrences to get the habitat info
df_occ_with_habitat = df_occurrences.merge(
    df_old_reserve_transects[['UID', 'Pre: Transect physical habitat']],
    how='inner',
    left_on='TransectUID',
    right_on='UID'
)

# Step 3: Group and count occurrences (MNI)
mni_counts = df_occ_with_habitat['Pre: Transect physical habitat'].value_counts()

# Step 4: Add MNI row to the summary table
summary_table.loc['MNI (n)'] = summary_table.columns.map(mni_counts).fillna(0).astype(int)

# By Taxa
# Filter transects to 'old reserve' only
df_occ_with_habitat['Taxon Label'] = df_occ_with_habitat['Post: Taxon guess?'].fillna(
    df_occ_with_habitat['Pre: Taxon']
)

taxon_summary = df_occ_with_habitat.pivot_table(
    index='Taxon Label',
    columns='Pre: Transect physical habitat',
    values='ID',
    aggfunc='count',
    fill_value=0
)

# Align and append to summary_table
taxon_summary = taxon_summary.reindex(columns=summary_table.columns, fill_value=0)
summary_table = pd.concat([summary_table, taxon_summary])
```

```python
# km2 covered
# Multiply each transect's distance by 0.05 to estimate area
df_old_reserve_transects['area_km2'] = df_old_reserve_transects['distance_km'] * 0.05

# Sum area per habitat type
area_by_habitat = df_old_reserve_transects.groupby("Pre: Transect physical habitat")["area_km2"].sum()

# Align with column order of summary_table
#area_row = summary_table.columns.map(area_by_habitat).fillna(0).round(2)
area_row = pd.Series(summary_table.columns.map(area_by_habitat)).fillna(0).astype(float).round(2)
# Now assign the row (note: index should match summary_table.columns)
area_row.index = summary_table.columns

# Add to summary_table
summary_table.loc['km2 covered'] = area_row
# summary_table.head()
```

```python
# Calculate MNI per kmÂ² (avoid division by zero just in case)
mni_row = summary_table.loc['MNI (n)']
area_row = summary_table.loc['km2 covered']

mni_per_km2 = (mni_row / area_row.replace(0, pd.NA)).round().fillna(0)
summary_table.loc['MNI/km2'] = pd.Series(mni_per_km2, index=summary_table.columns)
summary_table.loc['MNI/km2'].apply(type)
```

```python
# NISP (# of bones)
# Merge df_instances with df_occurrences (to get TransectUID)
df_instances_with_occ = df_instances.merge(
    df_occurrences[['ID', 'TransectUID']],
    how='left',
    left_on='OccurrenceID',
    right_on='ID',
    suffixes=('', '_occ')
)

# Merge with df_transects to get the habitat
df_instances_full = df_instances_with_occ.merge(
    df_transects[['UID', 'Pre: Transect physical habitat', 'Pre: On old reserve?']],
    how='left',
    left_on='TransectUID',
    right_on='UID',
    suffixes=('', '_transect')
)

# Filter to "old reserve" transects (if needed)

df_instances_old_reserve = df_instances_full[df_instances_full["Pre: On old reserve?"] == "Yes"].copy()

# Count instances per habitat (aka NISP)
nisp_counts = df_instances_old_reserve['Pre: Transect physical habitat'].value_counts()
nisp_row = summary_table.columns.map(nisp_counts).fillna(0).astype(int)
summary_table.loc['NISP (# of bones)'] = nisp_row
```

```python
# NISP/km2
# Get existing rows from the summary table
nisp_row = summary_table.loc['NISP (# of bones)']
area_row = summary_table.loc['km2 covered']

# Calculate NISP per kmÂ², round, and convert to int
nisp_per_km2 = (nisp_row / area_row.replace(0, pd.NA)).round().fillna(0).astype(int)

# Add to the summary table
summary_table.loc['NISP/km2'] = pd.Series(nisp_per_km2, index=summary_table.columns)
```

```python
# Weathering class percentages
df_briana_with_occ = df_briana_with_responses.merge(
    df_occurrences[['ID', 'TransectUID']],
    how='left',
    left_on='OccurrenceID',
    right_on='ID',
    suffixes=('', '_occ')
)

df_briana_with_occ_transect = df_briana_with_occ.merge(
    df_transects[['UID', 'Pre: Transect physical habitat', 'Pre: On old reserve?']],
    how='left',
    left_on='TransectUID',
    right_on='UID',
    suffixes=('', '_transect')
)

# Only old reserve
df_briana_old_reserve = df_briana_with_occ_transect[
    df_briana_with_occ_transect['Pre: On old reserve?'] == 'Yes'
].copy()

weathering_counts = df_briana_old_reserve.pivot_table(
    index='Weathering class',
    columns='Pre: Transect physical habitat',
    values='UID',
    aggfunc='count',
    fill_value=0
)

weathering_percent = weathering_counts.div(weathering_counts.sum(axis=0), axis=1) * 100
weathering_percent = weathering_percent.round(1)

# Reindex to match summary table column order
weathering_percent = weathering_percent.reindex(columns=summary_table.columns, fill_value=0)

# Add rows to summary table
for label, row in weathering_percent.iterrows():
    summary_table.loc[f'Weathering class {label} (%)'] = row
```

```python
# Count total instances per habitat (denominator)
nisp_total = df_briana_old_reserve['Pre: Transect physical habitat'].value_counts()

# Count buried instances per habitat (numerator)
buried = df_briana_old_reserve[df_briana_old_reserve['Buried?'] == 'Yes']
buried_counts = buried['Pre: Transect physical habitat'].value_counts()

# Calculate percentage
buried_percent = (buried_counts / nisp_total * 100).round(1)

# Align with summary table columns
buried_row = summary_table.columns.map(buried_percent).fillna(0)

# Add row to summary table
summary_table.loc['% buried bones'] = buried_row
```

```python
# % bones with any carnivore damage

# Instances with any carnivore damage
# Count where there is *any* non-null damage level
# Count only valid carnivore damage levels (as strings just in case)
valid_damage_levels = {'1', '2', '3', '4'}

has_valid_carn_damage = df_briana_old_reserve[
    df_briana_old_reserve['Carn Damage 1 Level'].astype(str).isin(valid_damage_levels)
]

carn_counts = has_valid_carn_damage['Pre: Transect physical habitat'].value_counts()

# Calculate % per habitat
carn_percent = (carn_counts / nisp_total * 100).round(1)

# Align with summary_table columns
carn_row = summary_table.columns.map(carn_percent).fillna(0)

summary_table.loc['% bones with any carnivore damage'] = carn_row
```

```python
df_transect_metadata = df_transects.copy()
df_transect_metadata['area_km2'] = df_transect_metadata['distance_km'] * 0.05
mni_per_transect = df_occurrences['TransectUID'].value_counts()
df_transect_metadata['mni_count'] = df_transect_metadata['UID'].map(mni_per_transect).fillna(0).astype(int)
nisp_per_transect = df_instances_full['TransectUID'].value_counts()
df_transect_metadata['nisp_count'] = df_transect_metadata['UID'].map(nisp_per_transect).fillna(0).astype(int)
```

```python
# DataFrames to Prepare & Save as .pkl
# Name	Description
# df_occurrences_with_taxon	Occurrences + habitat + clean taxon labels
# df_taxon_by_habitat	Taxon abundance matrix by habitat
# df_taxon_by_transect	Taxon abundance matrix by transect
# df_taxon_by_instance (optional)	Taxon occurrence by instance (if needed later)

# 1. Create a clean Taxon label column
df_occurrences['Taxon Label'] = df_occurrences['Post: Taxon guess?'].fillna(
    df_occurrences['Pre: Taxon']
)

# 2. Merge with transect metadata to get habitat

df_occurrences_with_taxon = df_occurrences.merge(
    df_transects[['UID', 'Pre: Transect physical habitat', 'Pre: On old reserve?']],
    how='left',
    left_on='TransectUID',
    right_on='UID'
)

# 3. Create composition matrices
# A. Taxon by Habitat
df_taxon_by_habitat = df_occurrences_with_taxon.pivot_table(
    index='Taxon Label',
    columns='Pre: Transect physical habitat',
    values='ID',
    aggfunc='count',
    fill_value=0
)

# B. Taxon by Transect
df_taxon_by_transect = df_occurrences_with_taxon.pivot_table(
    index='Taxon Label',
    columns='TransectUID',
    values='ID',
    aggfunc='count',
    fill_value=0
)

# 4. Save all to pickle files
import os
os.makedirs("data", exist_ok=True)

df_occurrences_with_taxon.to_pickle("../data/pkl/df_occurrences_with_taxon.pkl")
df_taxon_by_habitat.to_pickle("../data/pkl/df_taxon_by_habitat.pkl")
df_taxon_by_transect.to_pickle("../data/pkl/df_taxon_by_transect.pkl")
```
