```python
import pandas as pd
import numpy as np
import openpyxl
from pathlib import Path

df_transects = pd.read_pickle("../data/pkl/df_transects.pkl")
df_transects.head()
```

```python

# Weathering class lookup
weathering_file = Path("../data/import/excel/Weathering class.xlsx")
df_weathering = pd.read_excel(weathering_file)
# Mapping of weathering class to age ranges
weathering_lookup = df_weathering.set_index("Weathering class")[["Age min", "Age max"]]
```

```python
df_occurrences = pd.read_pickle("../data/pkl/df_occurrences.pkl")

hi = 'Post: Taxon Guess?'
lo = 'Post: Taxon guess?'

# (optional but recommended) treat blanks/NaN-like strings as real NaN
for c in (hi, lo):
    if c in df_occurrences.columns:
        df_occurrences[c] = (df_occurrences[c]
            .astype(str)
            .str.strip()
            .replace({'': np.nan, 'nan': np.nan, 'NaN': np.nan, 'None': np.nan, 'NULL': np.nan, 'null': np.nan}, regex=False)
        )

# Use lower-g values only where capital-G is null
if lo in df_occurrences.columns:
    df_occurrences[hi] = df_occurrences[hi].combine_first(df_occurrences[lo])  # or: .fillna(df_occurrences[lo])
    df_occurrences = df_occurrences.drop(columns=[lo])

df_occurrences.to_pickle("../data/pkl/df_occurrences.pkl")


df_occurrences.head(10)
```

```python
df_instances = pd.read_pickle("../data/pkl/df_instances.pkl")
df_instances.head()

df_briana_with_responses = pd.read_pickle("../data/pkl/df_briana_with_responses.pkl")
df_briana_with_responses.head()
```

```python
# Estimate year of death for instances and occurrences
inst_tmp = df_instances.merge(
    df_briana_with_responses[["UID", "Weathering class"]],
    on="UID",
    how="left",
).join(weathering_lookup, on="Weathering class")
inst_tmp["RecordingStartTime"] = pd.to_datetime(inst_tmp["RecordingStartTime"])

latest = inst_tmp["RecordingStartTime"].dt.year - inst_tmp["Age min"]
earliest = inst_tmp["RecordingStartTime"].dt.year - inst_tmp["Age max"]
mask = inst_tmp["Weathering class"].notna()

if "year_died_latest" not in df_instances.columns:
    df_instances["year_died_latest"] = pd.NA
if "year_died_earliest" not in df_instances.columns:
    df_instances["year_died_earliest"] = pd.NA

df_instances.loc[mask, "year_died_latest"] = np.floor(latest[mask]).astype("Int64")
df_instances.loc[mask, "year_died_earliest"] = np.floor(earliest[mask]).astype("Int64")

df_instances["year_died_latest"] = df_instances["year_died_latest"].astype("Int64")
df_instances["year_died_earliest"] = df_instances["year_died_earliest"].astype("Int64")

df_instances.to_pickle("../data/pkl/df_instances.pkl")

if "year_died_latest" not in df_occurrences.columns or "year_died_earliest" not in df_occurrences.columns:
    occ_years = (
        df_instances.groupby("OccurrenceID")[ ["year_died_latest", "year_died_earliest"] ]
        .min()
    )
    for col in occ_years.columns:
        occ_years[col] = pd.array(occ_years[col], dtype="Int64")
    df_occurrences = df_occurrences.merge(
        occ_years, left_on="ID", right_index=True, how="left",
    )

for col in ["year_died_latest", "year_died_earliest"]:
    if col in df_occurrences.columns:
        df_occurrences[col] = pd.array(
            np.floor(pd.to_numeric(df_occurrences[col], errors="coerce")),
            dtype="Int64",
        )

df_occurrences.to_pickle("../data/pkl/df_occurrences.pkl")

```

```python
# Estimate year of death for Briana responses
briana_tmp = df_briana_with_responses.merge(
    df_instances[["UID", "RecordingStartTime"]], on="UID", how="left",
)
briana_tmp = briana_tmp.join(weathering_lookup, on="Weathering class")
briana_tmp["RecordingStartTime"] = pd.to_datetime(briana_tmp["RecordingStartTime"])

latest = briana_tmp["RecordingStartTime"].dt.year - briana_tmp["Age min"]
earliest = briana_tmp["RecordingStartTime"].dt.year - briana_tmp["Age max"]
mask = briana_tmp["Weathering class"].notna()

if "year_died_latest" not in df_briana_with_responses.columns:
    df_briana_with_responses["year_died_latest"] = pd.NA
if "year_died_earliest" not in df_briana_with_responses.columns:
    df_briana_with_responses["year_died_earliest"] = pd.NA

idx = df_briana_with_responses.index[mask]
df_briana_with_responses.loc[idx, "year_died_latest"] = np.floor(latest[mask]).astype("Int64")
df_briana_with_responses.loc[idx, "year_died_earliest"] = np.floor(earliest[mask]).astype("Int64")

df_briana_with_responses["year_died_latest"] = df_briana_with_responses["year_died_latest"].astype("Int64")
df_briana_with_responses["year_died_earliest"] = df_briana_with_responses["year_died_earliest"].astype("Int64")

df_briana_with_responses.to_pickle("../data/pkl/df_briana_with_responses.pkl")
```

```python
df_fire_with_responses = pd.read_pickle("../data/pkl/df_fire_with_responses.pkl")
df_fire_with_responses.head()
```

```python
# NISP (# of bones)
# Merge df_instances with df_occurrences (to get TransectUID)
df_instances_with_occ = df_instances.merge(
    df_occurrences[['ID', 'TransectUID']],
    how='inner',
    left_on='OccurrenceID',
    right_on='ID',
    suffixes=('', '_occ')
)

# Merge with df_transects to get the habitat
df_instances_full = df_instances_with_occ.merge(
    df_transects[['UID', 'Pre: Transect physical habitat', 'Pre: On old reserve?']],
    how='inner',
    left_on='TransectUID',
    right_on='UID',
    suffixes=('', '_transect')
)
```

```python
df_transect_metadata = df_transects.copy()
df_transect_metadata['area_km2'] = df_transect_metadata['distance_km'] * 0.05
mni_per_transect = df_occurrences['TransectUID'].value_counts()
df_transect_metadata['mni_count'] = df_transect_metadata['UID'].map(mni_per_transect).fillna(0).astype(int)
nisp_per_transect = df_instances_full['TransectUID'].value_counts()
df_transect_metadata['nisp_count'] = df_transect_metadata['UID'].map(nisp_per_transect).fillna(0).astype(int)
```

```python
# 1. Create a clean Taxon label column
df_occurrences['Taxon Label'] = (
    df_occurrences['Post: Taxon Guess?']
    .fillna(df_occurrences['Pre: Taxon'])
)

df_occurrences['Taxon Label'] = df_occurrences['Taxon Label'].replace({
    '36;37': 'Bovidae (large)'
})
df_occurrences['Taxon Label'] = df_occurrences['Taxon Label'].replace({
    '43;41': 'Bovidae (small)'
})
df_occurrences['Taxon Label'] = df_occurrences['Taxon Label'].replace({
    '42;46': 'Bovidae (small)'
})

# 2. Merge with transect metadata to get habitat

df_occurrences_with_taxon = df_occurrences.merge(
    df_transects[['UID', 'Pre: Transect physical habitat', 'Pre: On old reserve?']],
    how='left',
    left_on='TransectUID',
    right_on='UID'
)
for col in ["year_died_latest", "year_died_earliest"]:
    if col in df_occurrences_with_taxon.columns:
        df_occurrences_with_taxon[col] = pd.array(
            np.floor(pd.to_numeric(df_occurrences_with_taxon[col], errors="coerce")),
            dtype="Int64",
        )

# 3. Create composition matrices
# A. Taxon by Habitat
df_taxon_by_habitat = df_occurrences_with_taxon.pivot_table(
    index='Taxon Label',
    columns='Pre: Transect physical habitat',
    values='ID',
    aggfunc='count',
    fill_value=0
)

# B. Taxon by Transect
df_taxon_by_transect = df_occurrences_with_taxon.pivot_table(
    index='Taxon Label',
    columns='TransectUID',
    values='ID',
    aggfunc='count',
    fill_value=0
)

# 4. Save all to pickle files
import os
os.makedirs("data", exist_ok=True)

df_occurrences_with_taxon.to_pickle("../data/pkl/df_occurrences_with_taxon.pkl")
df_taxon_by_habitat.to_pickle("../data/pkl/df_taxon_by_habitat.pkl")
df_taxon_by_transect.to_pickle("../data/pkl/df_taxon_by_transect.pkl")

```

```python
output_path = "../data/export/excel/source_data.xlsx"

instances_clean = df_instances.rename(columns=lambda c: c.strip())
occurrences_clean = df_occurrences.rename(columns=lambda c: c.strip())
briana_clean = df_briana_with_responses.rename(columns=lambda c: c.strip())

# normalize instance column names to ensure size-class fields are preserved
inst_col_map = {}
for col in instances_clean.columns:
    key = col.strip().lower()
    if key == "pre: size class":
        inst_col_map[col] = "Pre: Size class"
    elif key == "post: size class":
        inst_col_map[col] = "Post: Size Class"
    elif key == "pre: age":
        inst_col_map[col] = "Pre: Age"
instances_norm = instances_clean.rename(columns=inst_col_map)

inst_cols = ["UID", "Pre: Size class", "Pre: Age", "Post: Size Class"]
instances_subset = instances_norm.reindex(columns=inst_cols)

occ_cols = [
    "ID",
    "TransectUID",
    "Pre: Taxon",
    "Post: Taxon",
    "Taxon Label",
    "Post: Sex",
    "Pre: Sex",
    "Post: Notes on Occurence End",
]
occurrences_subset = occurrences_clean.reindex(columns=occ_cols)

briana_cols = [
    "UID",
    "OccurrenceID",
    "InstanceNumber",
    "Articulated?",
    "Complete?",
    "Innominate Portion",
    "Mandible Portion",
    "Rib Portion",
    "Scapula Portion",
    "Weathering class",
    "What element is this?",
]
briana_subset = briana_clean.reindex(columns=briana_cols)

briana_sheet = (
    briana_subset
    .merge(instances_subset, on="UID", how="left")
    .merge(
        occurrences_subset,
        left_on="OccurrenceID",
        right_on="ID",
        how="left",
    )
    .rename(
        columns={
            "TransectUID": "Transect number",
            "ID": "Occurrence id",
            "InstanceNumber": "Occurrence number",
            "Pre: Age": "Age",
        }
    )
    .reindex(
        columns=[
            "Transect number",
            "Occurrence id",
            "Occurrence number",
            "Pre: Taxon",
            "Post: Taxon",
            "Taxon Label",
            "Post: Sex",
            "Post: Size Class",
            "Age",
            "Pre: Sex",
            "Pre: Size class",
            "Post: Notes on Occurence End",
            "Articulated?",
            "Complete?",
            "Innominate Portion",
            "Mandible Portion",
            "Rib Portion",
            "Scapula Portion",
            "Weathering class",
            "What element is this?",
        ]
    )
)

with pd.ExcelWriter(output_path, engine='xlsxwriter') as writer:
    df_transects.to_excel(writer, sheet_name='Transects', index=False)
    df_occurrences.to_excel(writer, sheet_name='Occurrences', index=False)
    df_instances.to_excel(writer, sheet_name='Instances', index=False)
    df_briana_with_responses.to_excel(writer, sheet_name='Briana', index=False)
    df_fire_with_responses.to_excel(writer, sheet_name='Fire', index=False)
    df_transect_metadata.to_excel(writer, sheet_name='TransectData', index=False)
    briana_sheet.to_excel(writer, sheet_name='OccurrencesBriana', index=False)

print(f"✅ Excel saved to: {output_path}")
```
